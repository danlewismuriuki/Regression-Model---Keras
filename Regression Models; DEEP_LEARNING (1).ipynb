### Split data into predictors and target
The target variable in this problem is the concrete sample strength. Therefore, our predictors will be all the other columns, the independant variable.

```python
#split the data-frame into target(y) and predictors(X)

concrete_data_columns = concrete_data.columns

X = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] # all columns except Strength
y = concrete_data['Strength'] # Strength column
```

```python
#define the number of columns in our predictors which is the number of inputs of regression model

n_cols = len(X.columns)
n_cols
```

## A. Build a baseline model (5 marks)

### Let's go ahead and import the Keras library

```python
# import keras libraries 

import keras
```

```python
# import sci-kit learn library and train_test_split function

from keras.models import Sequential
from keras.layers import Dense
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
# define regression model

def regression_model():
    # create model
    model = Sequential()
    model.add(Dense(n_cols, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
#split the data randomly into test and train parts (test_size = 30%)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#create a regression model and store it in an object called model.

model = regression_model()
```

### Train and Test the Network

```python
#train the model using fit attribute of the model

model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=50)
```

```python
#check the  mean_squared_error of the model.

scores = model.evaluate(X_test, y_test, verbose=0)
```

```python
# repeat the last three steps 50 times and store the value of mean_squared_error for each loop in a list called mean.

mean = []
for i in range(50):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    model = regression_model()
    model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=50)
    score = model.evaluate(X_test, y_test, verbose=0)
    mean.append(score)
```

```python
# convert mean list to numpy array in order to calculate the mean and sd value of all 50 mean_squared_error.

mean_sqr_error = np.array(mean)
print('the mean value of 50 mean_squared_error of my model is {}'.format(np.mean(mean_sqr_error,)))
print('the standard deviation value of 50 mean_squared_error of my model is {} '.format(np.std(mean_sqr_error)))
```

### D. Increase the number of hidden layers (5 marks)

```python
#Define a new regression model which contains on three hidden layres (10 nodes and relu activation function)
def regression_model_1():
    # create model
    model = Sequential()
    model.add(Dense(n_cols, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```
